```python
def palindrome(s):
    rev = ""
    for i in s:
        rev = i + rev
    if s == rev:
        print("Palindrome")
    else:
        print("Not a Palindrome")

text = input("Enter a string: ")
palindrome(text)
```

```python
def palindrome(num):
    temp = num
    rev = 0

    while num > 0:
        digit = num % 10
        rev = rev * 10 + digit
        num = num // 10

    if temp == rev:
        print("Palindrome Number")
    else:
        print("Not a Palindrome Number")

n = int(input("Enter a number: "))
palindrome(n)
```

**Maps**

```python
#map(if,iterable)
import math

# Function to find area of circle
def area_circle(radius):
    return math.pi * radius * radius

# List of radii
radii = [2, 3, 4, 5]
```

```python
# Using map()
areas = list(map(area_circle, radii))

print("Circle Areas:", areas)
```

```python
# Function to increase salary by 10%
def increase_salary(salary):
    return round(salary * 1.10)

# List of employee salaries
salaries = [25000, 30000, 45000, 50000]

# Using map()
new_salaries = list(map(increase_salary, salaries))

print("Old Salaries:", salaries)
print("New Salaries:", new_salaries)
```

```python
#using lambda
# List of employee salaries
salaries = [25000, 30000, 45000, 50000]

# Increase salary by 10% and round to 2 decimal places
new_salaries = list(map(lambda salary: round(salary * 1.10), salaries))

print("Old Salaries:", salaries)
print("New Salaries:", new_salaries)
```

```python
# Function to convert Fahrenheit to Celsius
def fahrenheit_to_celsius(f):
    return (f - 32) * 5 / 9

# List of Fahrenheit temperatures
fahrenheit = [32, 68, 77, 98.6, 212]

# Using map() to convert
celsius = list(map(fahrenheit_to_celsius, fahrenheit))

print("Fahrenheit:", fahrenheit)
print("Celsius:", celsius)
```

```python
import math

def area_circle(radius):
    return math.pi * radius * radius

radii = [2, 3, 4, 5]

result = list(filter(lambda r: area_circle(r) > 10, radii))

for r in result:
    print("Radius:", r, "Area:", area_circle(r))
```

```python
# Function to check marks
def passed(mark):
    return mark >= 60

# List of student marks
marks = [45, 78, 56, 90, 60, 38, 72]

# Using filter()
passed_students = list(filter(passed, marks))

print("Student Marks:", marks)
print("Marks 60 and above:", passed_students)
```

```python
#using lambda
marks = [45, 78, 56, 90, 60, 38, 72]

passed_students = list(filter(lambda mark: mark >= 60, marks))

print(passed_students)
```

```python
#reduce
from functools import reduce
import math

# Function to find area of circle
def area_circle(radius):
    return math.pi * radius * radius

# List of radii
radii = [2, 3, 4, 5]

# Find total area using reduce()
total_area = reduce(lambda x, y: x + area_circle(y), radii, 0)

print("Total Area =", total_area)
```

```python
from functools import reduce

# List of monthly sales
sales = [1200, 1500, 2100, 1800]

# Using reduce() to find total sales
total_sales = reduce(lambda x, y: x + y, sales)

print("Monthly Sales:", sales)
print("Total Sales:", total_sales)
```

```python
#map and reduce
from functools import reduce
import math

# List of radii
radii = [2, 3, 4, 5]

# Calculate areas using map()
areas = list(map(lambda r: math.pi * r * r, radii))

# Sum all areas using reduce()
total = reduce(lambda x, y: x + y, areas)

print("Areas:", areas)
print("Total Area:", total)
```

```python
# List of numbers
numbers = [2, 5, 8, 11, 14, 17]

# Filter even numbers
even_numbers = filter(lambda x: x % 2 == 0, numbers)

# Square the even numbers
squared_numbers = list(map(lambda x: x * x, even_numbers))

print("Squared Even Numbers:", squared_numbers)
```

```python
from functools import reduce

# List of employee salaries
salaries = [2500, 3200, 4500, 2800, 5000]

# Filter salaries greater than 3000
filtered = filter(lambda x: x > 3000, salaries)

# Increase each salary by 5%
updated = list(map(lambda x: x * 1.05, filtered))

# Find total salary
total = reduce(lambda x, y: x + y, updated)

print("Updated Salaries:", updated)
print("Total Salary:", total)
```
